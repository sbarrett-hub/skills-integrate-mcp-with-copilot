# Configuration Setup Verification

This notebook checks if the configuration setup for the project is correct by verifying the presence, parsing, and validation of configuration files, and testing them in the application context.

## 1. Verify Configuration Files Exist

Check for the presence of required configuration files (e.g., `.vscode/mcp.json`, `config.yaml`, `.env`).

```python
import os

required_files = ['.vscode/mcp.json', 'config.yaml', '.env']
missing_files = [f for f in required_files if not os.path.exists(f)]

print("Required configuration files:")
for f in required_files:
    print(f"- {f}: {'FOUND' if os.path.exists(f) else 'MISSING'}")

if missing_files:
    print("\nMissing files:", missing_files)
else:
    print("\nAll required configuration files are present.")
```

## 2. Load and Parse Configuration

Attempt to load and parse the configuration files. If a file is missing, skip its parsing.

```python
import json

# Load .vscode/mcp.json if present
mcp_config = None
if os.path.exists('.vscode/mcp.json'):
    with open('.vscode/mcp.json') as f:
        try:
            mcp_config = json.load(f)
            print("Loaded .vscode/mcp.json:")
            print(json.dumps(mcp_config, indent=2))
        except Exception as e:
            print("Error parsing .vscode/mcp.json:", e)
else:
    print(".vscode/mcp.json not found.")

# Load config.yaml if present
try:
    import yaml
except ImportError:
    yaml = None
    print("PyYAML not installed. Skipping YAML parsing.")

config_yaml = None
if yaml and os.path.exists('config.yaml'):
    with open('config.yaml') as f:
        try:
            config_yaml = yaml.safe_load(f)
            print("Loaded config.yaml:")
            print(config_yaml)
        except Exception as e:
            print("Error parsing config.yaml:", e)
else:
    print("config.yaml not found or PyYAML missing.")

# Load .env if present
try:
    from dotenv import dotenv_values
except ImportError:
    dotenv_values = None
    print("python-dotenv not installed. Skipping .env parsing.")

env_config = None
if dotenv_values and os.path.exists('.env'):
    env_config = dotenv_values('.env')
    print("Loaded .env:")
    print(env_config)
else:
    print(".env not found or python-dotenv missing.")
```

## 3. Validate Configuration Parameters

Check that all necessary configuration parameters are present and have valid values. Print out missing or invalid parameters.

```python
# Example validation for .vscode/mcp.json
required_mcp_keys = ["servers"]
missing_keys = []

if mcp_config:
    for key in required_mcp_keys:
        if key not in mcp_config:
            missing_keys.append(key)
    if missing_keys:
        print("Missing keys in .vscode/mcp.json:", missing_keys)
    else:
        print("All required keys present in .vscode/mcp.json.")
else:
    print("Cannot validate .vscode/mcp.json (not loaded).")

# Add similar validation for config.yaml and .env if needed
```

## 4. Test Configuration in Application Context

Attempt to use the configuration to initialize a core component of the application (e.g., database connection, API client) and report success or failure.

```python
# Example: Test if MCP server URL is reachable (for demonstration)
import requests

if mcp_config and "servers" in mcp_config and "github" in mcp_config["servers"]:
    url = mcp_config["servers"]["github"].get("url")
    if url:
        try:
            response = requests.get(url)
            print(f"Tried to reach {url}. Status code: {response.status_code}")
        except Exception as e:
            print(f"Failed to reach {url}: {e}")
    else:
        print("No URL found in MCP config for GitHub server.")
else:
    print("MCP config for GitHub server not found.")
```
